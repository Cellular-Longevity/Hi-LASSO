# Getting Started

## Data load

```python
import pandas as pd 
X = pd.read_csv('simulation_data_x.csv')
y = pd.read_csv('simulation_data_y.csv')
```

```python
X.head()
```

```python
y.head()
```

## General Usage

```python
from hi_lasso.hi_lasso import HiLasso
hi_lasso = HiLasso(X, y, q1='auto', q2='auto', B='auto', d=0.05, alpha=0.95, par_opt=False) 
fitted_hi_lasso = hi_lasso.fit()
```

```python
fitted_hi_lasso.coef_
```

```python
fitted_hi_lasso.p_values_
```

```python
fitted_hi_lasso.selected_var_
```

## Parallel Processing Usage

If `par_opt` is set to **True**, parallel processing can be used for bootstrapping.

You can specify `max_workers`(the number of cores) to use in parallel processing.

```python
hi_lasso = HiLasso(X, y, par_opt=True, max_workers=12) 
fitted_hi_lasso = hi_lasso.fit()
```

```python
fitted_hi_lasso.coef_
```

```python
fitted_hi_lasso.p_values_
```

```python
fitted_hi_lasso.selected_var_
```
